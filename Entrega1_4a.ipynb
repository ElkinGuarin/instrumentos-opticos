```python
#Librerías
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Lectura los archivos en arreglos complejos. Antes se editaron y se cambió la i por la j.
txt = np.loadtxt("a.txt", delimiter = ',', dtype = 'complex')
csv = np.loadtxt("a.csv", delimiter = ',', dtype = 'complex')
dat = np.loadtxt("a.dat", delimiter = ',', dtype = 'complex')
```

```python
#Se grafican las componentes del archivo a.txt
fig, a_txt = plt.subplots(1,4,figsize=(15, 15))
a_txt[0].set_title("Parte real")
a_txt[0].imshow(np.real(txt))
a_txt[1].set_title("Parte Imaginaria")
a_txt[1].imshow(np.imag(txt))
a_txt[2].set_title("Fase")
a_txt[2].imshow(np.angle(txt))
a_txt[3].set_title("Superposición")
a_txt[3].imshow(np.angle(txt)+np.imag(txt)+np.real(txt),cmap="gray")
fig.tight_layout()
plt.show()
```
