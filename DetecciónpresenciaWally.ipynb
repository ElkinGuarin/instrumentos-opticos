```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_rgb = cv2.imread('c.jpg')
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
template = cv2.imread('c_clue.jpg',0)
#saves the width and height of the template into 'w' and 'h'
w, h = template.shape[::-1]
```

```python
res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.8
```

```python
loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    #draw rectangle on places where it exceeds threshold
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (255,255,255), 2)
cv2.imwrite('c_found.jpg',img_rgb)
```

```python
plt.figure()
plt.imshow(template)
#plt.colorbar()
plt.show()
```
